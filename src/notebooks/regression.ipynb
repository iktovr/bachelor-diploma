```python
from sindy.utils.plot import plot_regression_scores, plot_coef
```

```python
def plot_result(model, data, target):
    fig = plt.figure(figsize=(11, 5))
    axes = fig.subplots(1, 3, width_ratios=(0.49, 0.49, 0.02))
    axes[0].set_aspect('equal', 'box')
    plot_regression_scores(model, data, target, ax=axes[0])
    plot_coef(model, ax=axes[1], cbar_ax=axes[2])
    fig.subplots_adjust(wspace=0.1)
    plt.show()
```

```python
data, target, true_coef = make_regression(n_samples=100, n_features=100, n_informative=10, noise=5, coef=True)
```

```python
plot_coef(true_coef, figsize=(6, 5))
```

```python
lin = LinearRegression(fit_intercept=False)
lin.fit(data, target)
plot_result(lin, data, target)
```

```python
ridge = Ridge(fit_intercept=False)
ridge.fit(data, target)
plot_result(ridge, data, target)
```

```python
lasso = Lasso(fit_intercept=False)
lasso.fit(data, target)
plot_result(lasso, data, target)
```
